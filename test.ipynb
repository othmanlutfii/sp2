```python
def var_naik_turun(df,list_dep):
    list_dep = list_dep.copy()
    data_terahir = df['Close'][-1]
    list_dep.insert(0, data_terahir)

    trend_list = []  # List baru untuk menyimpan hasil perulangan

    for i in range(len(list_dep)-1):
        if list_dep[i+1] < list_dep[i]:
            trend_list.append("DOWN")
        elif list_dep[i+1] >= list_dep[i]:
            trend_list.append("UP")
        else:
            trend_list.append("no change")
    return(trend_list)

    
```

```python
def banding_list(a, b, c):
    com_list = []
    for i in range(len(a)):
        if (a[i] == "UP" and b[i] == "UP") :
            com_list.append("UP")
        elif (b[i] == "UP" and c[i] == "UP"):
            com_list.append("UP")
        elif (a[i] == "UP" and c[i] == "UP"):
            com_list.append("UP")
        elif (a[i] == "UP" and b[i] == "UP"):
            com_list.append("UP")
        elif (a[i] == "UP" and b[i] == "UP" and c[i] == "UP"):
            com_list.append("UP")
        else:
            com_list.append("DOWN")
    return(com_list)
    
```

```python
def compare_lists(list1, list2, list3):
    comparison_list = []

    for i in range(len(list1)):
        count_up = 0
        count_down = 0

        if list1[i] == "up":
            count_up += 1
        elif list1[i] == "down":
            count_down += 1

        if list2[i] == "up":
            count_up += 1
        elif list2[i] == "down":
            count_down += 1

        if list3[i] == "up":
            count_up += 1
        elif list3[i] == "down":
            count_down += 1

        if count_up >= 2:
            comparison_list.append("up")
        elif count_down >= 2:
            comparison_list.append("down")
        else:
            comparison_list.append("no change")

    return comparison_list
```

```python
def merge_final(a,e,b,c,d):    

    import pandas as pd

    # Example lists
    dates = a
    list_act = e
    list1 = b
    list2 = c
    list3 = d

    # Create a dictionary with lists as values
    data = {'Date': dates,'Nilai_asli': list_act, 'Model_lstm': list1, 'model_arima': list2, 'model_svr': list3}

    # Create a DataFrame from the dictionary
    df = pd.DataFrame(data)

    # Convert 'Date' column to datetime type
    df['Date'] = pd.to_datetime(df['Date'], format="%d-%m-%Y")

    # Set 'Date' as the index
    df.set_index('Date', inplace=True)
    return(df)
```

```python
# Plot the data
import matplotlib.pyplot as plt
df = merge_final(dt,pred,pred_lstm,pred_arima,pred_svr)
df
```
